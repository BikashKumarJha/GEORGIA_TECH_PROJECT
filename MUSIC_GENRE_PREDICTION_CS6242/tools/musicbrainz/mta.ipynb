```python
import musicbrainzngs
import pandas as pd
musicbrainzngs.set_useragent("mta","0.1",contact="spothana3@gatech.edu")
result = musicbrainzngs.search_artists(country="US", limit=200, offset=0)
artist_df = pd.DataFrame.from_dict(result['artist-list'])
artist_df.head()
```

```python

result = musicbrainzngs.search_labels(country="US",limit=10, offset=0)
label_df = pd.DataFrame.from_dict(result['label-list'])
label_df.head()
```

```python
result = musicbrainzngs.search_recordings(country="US",limit=10, offset=0)
recording_df = pd.DataFrame.from_dict(result['recording-list'])
recording_df.head()
```
